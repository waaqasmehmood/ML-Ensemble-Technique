```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# loading data into pandas data frame

heart_data = pd.read_csv("./heart.csv")



plt.figure(figsize=(7, 7))  # Set figure size
sns.set(style="whitegrid")  # Set style to whitegrid for better readability

# Create the count plot
ax = sns.countplot(x='target', data=heart_data, palette='viridis')

# Set IEEE style guidelines
ax.set_title('Count Plot of Target Variable', fontsize=14, weight='bold')
ax.set_xlabel('Target Variable', fontsize=12)
ax.set_ylabel('Count', fontsize=12)

# Set ticks and labels size
ax.tick_params(axis='both', which='major', labelsize=10)

# Save the plot in IEEE format (e.g., EPS format)
plt.savefig('CountPlotOfTarget.png', format='png', dpi=500, bbox_inches='tight')

# Show the plot
plt.show()
```
