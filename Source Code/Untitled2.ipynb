```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import statsmodels.api as sm
heart_data = pd.read_csv("./heart.csv")
data = heart_data

plt.figure(figsize=(7, 7))  # Set figure size
sns.set(style="whitegrid")  # Set style to whitegrid for better readability

# Create the Q-Q plot
sm.qqplot(data, line='45', fit=True)
plt.title('Q-Q Plot', fontsize=14, weight='bold')
plt.xlabel('Theoretical Quantiles', fontsize=12)
plt.ylabel('Sample Quantiles', fontsize=12)
plt.grid(True)

# Save the plot in PNG format
plt.savefig('qq_plot_ieee.png', format='png', dpi=500, bbox_inches='tight')

# Show the plot
plt.show()
```
